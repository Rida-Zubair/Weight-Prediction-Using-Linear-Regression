```python


# Load the CSV file into a DataFrame
data = pd.read_csv(processed_file_path)

# Display the first few rows to confirm successful loading
print(data.head(3))
```

```python
# Drop the 'Height' column
# Assuming 'data' is your DataFrame
# Replace Gender strings with numerical values (Female -> 1, Male -> 0)
data['Gender'] = data['Gender'].map({'Female': 1, 'Male': 0})
data['Genetics'] = data['Genetics'].map({'Yes': 1, 'No': 0})
# Function to convert height from ft.inch to cm
def convert_height_to_cm(height):
    if isinstance(height, (float, int)):  # Handle numbers directly
        feet = int(height)  # Extract feet
        inches = (height - feet) * 12  # Convert fractional feet to inches
        return feet * 30.48 + inches * 2.54  # Convert to cm
    return None


# Apply the height conversion
data['Height_cm'] = data['Height'].apply(convert_height_to_cm)
print(data[['Height', 'Height_cm']])


# Function to map sleep duration strings to integers
def map_sleep_duration(sleep):
    sleep_mapping = {
        'Less than 5 hours': 4,
        '5-6 hours': 5,
        '7-8 hours': 7,
        'More than 8 hours': 9
    }
    return sleep_mapping.get(sleep, None)  # Default to None for invalid entries


# Apply the sleep duration mapping
data['Sleep_Duration'] = data['Sleep_Duration'].apply(map_sleep_duration)
data = data.drop(columns=['Height'])

# Convert 'Exercise per Week' into integer (lower boundary of hours)
exercise_mapping = {
    '0 hours': 0,
    '1-2 hours': 1,
    '3-4 hours': 3,
    '5-6 hours': 5,
    '6-7 times': 6,  # Added for full mapping
    '7+ hours': 7
}
data['Exercise per Week'] = data['Exercise per Week'].map(exercise_mapping)

# Convert 'Medical History' into integers
medical_mapping = {
    'No diagnosed conditions': 0,
    'Other': 1
}
data['Medical History'] = data['Medical History'].map(medical_mapping)

# Convert 'Genetics' into integers
genetics_mapping = {
    'No': 0,
    'Yes': 1
}
data['Genetics'] = data['Genetics'].map(genetics_mapping)

# Display the first few rows of the updated data
print(data.head())

# Save the updated data to a new CSV file
processed_file_path = 'processed_weight_final1.csv'
data.to_csv(processed_file_path, index=False)

print(f"Updated data saved to {processed_file_path}")
```
